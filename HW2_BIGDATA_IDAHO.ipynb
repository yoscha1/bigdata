```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
df = pd.read_csv('HW2_Housing Survey.csv')
```

```python
print(df)
```

```python
countrows = df[df['VAL'] >= 14].shape[0]
print('Number of properties worth more than $1,000,000 =', countrows)
```

```python
persons = df['NP']
print(persons)
```

```python
sumpersons = df['NP'].sum()
print('Sum =',sumpersons)
avgpersons = sumpersons/df['NP'].shape[0]
print('Average number of persons in a household =', avgpersons)
```

```python
print(df['NP'].shape[0])
```

```python
correl = df[['VAL', 'NP']]
print(correl)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

df2 = correl

plt.figure(figsize=(8, 6))
sns.scatterplot(x='VAL', y='NP', data=df2)
plt.title('Scatter Plot of VAL vs NP')
plt.xlabel('VAL')
plt.ylabel('NP')
plt.show()
```

```python
groupedcorrel = df2.groupby('VAL').mean().reset_index()
groupedcorrel.plot(x='NP', y='VAL', kind='scatter', figsize=(8, 8))
correl.corr()
```

```python
faminc = df['FINCP'].fillna(0).astype(int)
faminc = faminc - faminc.min()
print(faminc)

faminc.hist(bins=30, color='red', edgecolor='black')
plt.xlabel('Family Incomes')
plt.ylabel('Count')
plt.title('Family Incomes (Before Normalization)')
```

```python
faminc = df['FINCP'].fillna(0).astype(int)
faminc = faminc - faminc.min()
min_val = faminc.min()
max_val = faminc.max()

# Perform Min-Max scaling
normalized_faminc = (faminc - min_val) / (max_val - min_val)
print(normalized_faminc)

normalized_faminc.hist(bins=30, color='skyblue', edgecolor='black')
plt.xlabel('Normalized Family Incomes')
plt.ylabel('Count')
plt.title('Family Incomes (After Normalization)')
```
