```python
from neuralprophet import NeuralProphet
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

movies = pd.read_csv("movies.csv")
movies[['director_name', 'imdb_score', 'movie_facebook_likes']]

# plt.plot(movies['movie_facebook_likes'], movies['imdb_score'], label = 'actual', c='g')
# plt.show()
movies.dropna(inplace=True)
```

```python
from sklearn.model_selection import train_test_split

X = movies.drop(['imdb_score'], axis=1)
y = movies['imdb_score']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
train_data = X_train.join(y_train)
train_data
```

```python
train_data.hist(figsize=(15,10))
```

```python
# correlation data
train_data = train_data.drop('director_name', axis=0)
train_data
# train_data.corr()
```

```python
import seaborn as sns
plt.figure(figsize=(15,10))
sns.heatmap(train_data.corr(), annot=True, cmap="YlGnBu")
```

```python
# taking right skewed category to take log and see distribution
train_data['director_facebook_likes'] = np.log(train_data['director_facebook_likes'] + 1)
train_data['actor_3_facebook_likes'] = np.log(train_data['actor_3_facebook_likes'] + 1)
train_data['actor_2_facebook_likes'] = np.log(train_data['actor_2_facebook_likes'] + 1)
train_data['actor_1_facebook_likes'] = np.log(train_data['actor_1_facebook_likes'] + 1)
train_data['movie_facebook_likes'] = np.log(train_data['movie_facebook_likes'] + 1)
train_data['gross'] = np.log(train_data['gross'] + 1)
train_data['num_voted_users'] = np.log(train_data['num_voted_users'] + 1)
train_data['cast_total_facebook_likes'] = np.log(train_data['cast_total_facebook_likes'] + 1)
train_data['facenumber_in_poster'] = np.log(train_data['facenumber_in_poster'] + 1)
train_data['num_user_for_reviews'] = np.log(train_data['num_user_for_reviews'] + 1)
train_data['budget'] = np.log(train_data['budget'] + 1)
train_data['aspect_ratio'] = np.log(train_data['aspect_ratio'] + 1)

train_data.hist(figsize=(15,10))
```
